DATA LOADING 


```python
#loading necessary libraries
import pandas as pd
```

```python
# loading the data
tf_full = pd.read_csv("transformed/transformed_full.csv")
tf_incremented = pd.read_csv("transformed/transformed_incremental.csv")
```

```python
#loading the data into parquet
tf_full.to_parquet('loaded/full_data.parquet')
tf_incremented.to_parquet('loaded/incremental_data.parquet')

```

```python
# Preview of the full transformed dataset
pd.read_parquet('loaded/full_data.parquet').head()
```
